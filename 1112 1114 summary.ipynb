```python
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.axes_grid1 import make_axes_locatable
import IPython.display as ipd
import numpy as np
%matplotlib notebook
from scipy.signal import lfilter
```

```python
amp = 2
sr = 10000       
dur = 0.5  
freq = 400.0
```

```python
t = np.arange(1, sr * dur+1)/sr
```

```python
theta = t * 2*np.pi * freq
```

```python
s = amp*np.sin(theta)    #sine 그래프에 진폭을 입력해주는 것.
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(t[0:1000], s[0:1000], '.')
ax.set_xlabel('time (s)')
ax.set_ylabel('real')
```

진폭을 곱해줌으로써 범위를 -1~1에서 -2~2로 확장시킴.

```python
c = amp*np.exp(theta*1j)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(t[0:100], c.real[0:100], c.imag[0:100], '.')
ax.set_xlabel('time (s)')
ax.set_ylabel('real')
ax.set_zlabel('imag')
```

```python
ipd.Audio(c.real, rate=sr)
```

최소의 sr을 사용하여 최대의 freq를 표현하려면 그 최대의 freq는 sr의 절반이다.

예를 들어 sr = 10hz라면 최대의 freq는 5hz이다. 이 최대의 freq는 nyquist freq이라고 한다.

cd의 음질은 44100인데, 이것의 nyquist freq는 22050이다. cd의 음질이 44100hz인 이유는 그 음질이 사람이 구별할 수 있는 최대의 소리이기 때문.

하지만, 사람의 음색에 대한 것은 그 이상의 주파수에 있기 때문에 유선전화를 썼을 때 구분이 애매함.

높은 주파수의 소리, 예를 들어 돌고래나 박쥐의 소리를 녹음하고 싶다면 그 소리의 두배에 해당하는 sr이 있어야함.

```python
F0 = 100; Fend = int(sr/2); s = np.zeros(len(t));   #0들을 만들어서 더해질 준비함.
for freq in range(F0, Fend+1, F0):     #F0부터 Fend까지 F0만큼의 간격으로
    theta = t * 2*np.pi * freq
    s = s + amp *np.sin(theta)
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(t[0:1000], s[0:1000]); 
ax.set_xlabel('time (msec)')
ipd.Audio(s, rate=sr)
```

```python
F0 = 100; Fend = int(sr/2); s = np.zeros(len(t));
for freq in range(F0, Fend+1, F0):
    theta = t * 2*np.pi * freq
    tmp = amp * np.sin(theta)
    s = s + tmp
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(t[0:1000], s[0:1000]); 
ax.set_xlabel('time (msec)')
ipd.Audio(s, rate=sr)
```

```python
def hz2w(F, sr):
    NyFreq = sr/2;
    w = F/NyFreq *np.pi;
    return w

def resonance (srate, F, BW):
    a2 = np.exp(-hz2w(BW,srate))
    omega = F*2*np.pi/srate
    a1 = -2*np.sqrt(a2)*np.cos(omega)
    a = np.array([1, a1, a2])
    b = np.array([sum(a)])
    return a, b
```

함수를 만드는 방법과 관련된 것.

def를 쓴 후 만들고 싶은 함수의 이름을 쓰고, ( , , )의 형태로 넣고 싶은 것 넣음.

return: 출력값. 하나일 필요는 없음.

```python
RG = 0  #RG: freq 값을 설정. 산맥의 위치. 산맥의 꼭대기를 기준으로 함.
BWG = 100 
a, b= resonance(sr,RG,BWG)   #resonance: 산맥이 어디에 있을 지 설정하는 것.  BWG: 산맥의 폭을 설정
s = lfilter(b, a, s, axis=0)
ipd.Audio(s, rate=sr)
```
